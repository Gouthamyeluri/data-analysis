# standardizing the data

```python
from sklearn.preprocessing import StandardScaler
s_data=StandardScaler().fit_transform(d)
s_data.shape
```

# co-variance matrix

```python
sample_data=s_data
co_var_matix=np.matmul(sample_data.T,sample_data)
co_var_matix.shape
```

# eigen values and vectors

```python
from scipy.linalg import eigh
values,vectors=eigh(co_var_matix,eigvals=(782,783))
vectors=vectors.T
```

```python
vectors.shape
```

```python
values2,vectors2=eigh(co_var_matix)
```

```python
vector3=vectors2[:,-3:]
```

```python
vector3=vector3.T
```

```python
sample_data.T.shape
```

```python
value3=values2[-2:]

sum(value3)/sum(values2)*100
```
